```python
def get_nr_of_sense_long(word):
  nrsense=0
  for synset in wn.synsets(word, pos=wn.NOUN):
    all_examples = ','.join([ f'"{e}"' for e in synset.examples() ])
    for l in synset.lemmas():
        # Find the sense number of the lemma of the target synset.
        # For the first we find all synsets for this lemma and 
        # in this synset list we find the position+1 of the target synset 
        all_synsets_of_l = wn.synsets(l.name())
        sense_num = all_synsets_of_l.index(l.synset()) + 1
        if l.name() == word:
          print(f"\t{l.name()}#{sense_num}")
```

```python
def get_nr_of_sense(word, t):
    return len(wn.synsets(word, pos=t))
```

```python
def find_noun(sentence, sentence_parse):
    splitted= sentence_parse.split()
    for i in range(len(splitted)):
        if splitted[i]== '(NN' or splitted[i]== '(NNS' or splitted[i]== '(NNP' or splitted[i]== '(NNPS':
            noun = re.sub('[()]', '', splitted[i+1])
            if noun in nouns_with_one_sense or noun in noun_sense_dict.keys():
                conjugation = re.sub('[()]', '', splitted[i])
                sentence = re.sub(fr'\b{noun}\b', "{noun}", sentence)
                return (True, sentence, noun, conjugation)

    return (False, 'no noun found', 'none', 'none')
```

```python
def find_verb(sentence, sentence_parse):
    splitted= sentence_parse.split()
    for i in range(len(splitted)):
        if splitted[i]== '(VB' or splitted[i]== '(VBD' or splitted[i]== '(VBG' or splitted[i]== '(VBN' or splitted[i]== '(VBP' or splitted[i]== '(VBZ':
            verb = re.sub('[()]', '', splitted[i+1])
            if verb in verbs_with_one_sense or verb in verb_sense_dict.keys():
                conjugation = re.sub('[()]', '', splitted[i])
                sentence = re.sub(fr'\b{verb}\b', "{verb}", sentence)
                return (True, sentence, verb, conjugation) # verandert alleen de eerste verb, fix

    return (False, 'no verb found', 'none', 'none')
```

```python
def find_adj(sentence, sentence_parse):
    splitted= sentence_parse.split()
    for i in range(len(splitted)):
        if splitted[i]== '(JJ' or splitted[i]== '(JJR' or splitted[i]== '(JJS':
            adj = re.sub('[()]', '', splitted[i+1])
            if adj in adjs_with_one_sense or adj in adj_sense_dict.keys():
                conjugation = re.sub('[()]', '', splitted[i])
                sentence = re.sub(fr'\b{adj}\b', "{adj}", sentence)
                return (True, sentence, adj, conjugation) # verandert alleen de eerste verb, fix

    return (False, 'no adj found', 'none', 'none')
```

```python
def traverse_tree(tree):
    np_trees=[]
    for subtree in tree:
        if type(subtree) == nltk.tree.Tree:
            np_list = traverse_tree(subtree)
            np_trees =np_trees +np_list
            if subtree.label()=="NP":
                np_trees.append(subtree)
            
    return np_trees
```

```python
AND = {'all', 'every', 'each', 'and'}
DIS = {'some', 'several', 'or',  'a'} 
NEG = {'no', 'not', 'neither', 'without'}
DEF = {'both'}
QUV = {'many', 'few'} 
IMP = {'if', 'when', 'unless'}


def Quantity_change(sentence, sentence_parse):
    
    parse_tree = Tree.fromstring(sentence_parse)
    np_trees = traverse_tree(parse_tree)
    
    for i in range(len(np_trees)):
        leaves =np_trees[i].leaves()
    # we will only do it for the first tree so only quantifiers at the beginning of a sentence will change
        for l in leaves:
            if l.lower() in AND:
                if l.lower()== 'and':
                    sentence = re.sub(fr'\b{l}\b', 'or', sentence, 1)
                    return sentence
                else:
                    sentence = re.sub(fr'\b{l}\b', 'no', sentence, 1)
                    return sentence
                    
            elif l.lower() in DIS:
                if l.lower()== 'or':
                    sentence = re.sub(fr'\b{l}\b', 'and', sentence, 1)
                    return sentence
                else:
                    sentence = re.sub(fr'\b{l}\b', 'no', sentence, 1)
                    return sentence
                    
                    
            elif l.lower() in QUV:
                sentence = re.sub(fr'\b{l}\b', 'no', sentence, 1)
                return sentence
                
            elif l.lower()== 'without':
                sentence = re.sub(fr'\b{l}\b', 'with', sentence, 1)
                return sentence
                
            elif l.lower()== 'neither':
                sentence = re.sub(fr'\b{l}\b', 'both', sentence, 1)
                return sentence
                
            elif l.lower()== 'no':
                sentence = re.sub(fr'\b{l}\b', 'all', sentence, 1)
                noun_list = []
                for n in leaves:
                    if n in nounSet:
                        noun_list.append(n)
                if  noun_list !=[]:
                    n = noun_list[-1]                  #  fix, you don't always want to change the last word of the tree
                    if pluralize(n)!=None:
                        sentence = sentence.replace(n, pluralize(n), 1)
                        return sentence

    return sentence
```

```python
def verb_past_to_future(sentence, sentence_parse):

    splitted= sentence_parse.split()
    sentsplit = sentence.split()
    
    for i in range(len(splitted)):
        if splitted[i]== '(VB' or splitted[i]== '(VBD' or splitted[i]== '(VBG' or splitted[i]== '(VBN' or splitted[i]== '(VBP' or splitted[i]== '(VBZ':
            verb = re.sub('[()]', '', splitted[i+1])
            conjugation = re.sub('[()]', '', splitted[i])
            
            if conjugation == 'VBD':
                new_verb= conjugate(verb= verb,tense = 'part') 
                for i in range(len(sentsplit)):
                    if sentsplit[i] == verb:
                        sentsplit[i]=new_verb
#                         if sentsplit[i-1] in aux_verbs:
#                             remove.append(sentsplit[i-1])

            elif conjugation == 'VBG':
                new_verb= conjugate(verb= verb,tense = 'p') 
                for i in range(len(sentsplit)):
                    if sentsplit[i] == verb:
                        sentsplit[i]=new_verb
                        if sentsplit[i-1] in aux_verbs:
                            sentsplit[i-1]=''
                            
            elif conjugation == 'VBN':
                new_verb= conjugate(verb= verb,tense = 'part')
                for i in range(len(sentsplit)):
                    if sentsplit[i] == verb:
                        sentsplit[i]=new_verb
                        if sentsplit[i-1] in aux_verbs:
                            sentsplit[i-1]=''
            
            sentence = ' '.join(sentsplit)
            sentence = sentence.replace('  ', ' ')
            
    return (sentence, sentence_parse)
```

```python
def hypothesis_pertubation(train_hypotheses, train_premises, fromm, to, verb_pertubation, noun_pertubation, list_operations_go):
    
    old_premises=[]
    new_hypotheses =[]
    hyp=[]
    
    iter= fromm
    while iter < to:
      # negated keeps track negation so it doens't swap with anto and negate, because that will give neutral
      # ent keeps track of entailment
        negated=False
        ent=False
        sentence = train_hypotheses[iter][0]
        sentence_parse = train_hypotheses[iter][1]
        sstart = sentence

        if random.random() < 0.3:
            if list_operations_go[0]==True:
                    new_sentence = hypo_and_other_premise(sentence, iter)
                    if new_sentence != sentence:
                        sentence= new_sentence
                        ent = True

        if verb_pertubation != '' and ent==False:
            tple = find_verb(sentence, sentence_parse) #(True, sentence, verb, conjugation)
            if tple[0]:
                  new_sentence = verb_pertubation(tple[2], tple[3], tple[1], sentence_parse)
                  if new_sentence!=sentence:
                      sentence=new_sentence
                      if verb_pertubation =='make_verbList_with_anto':
                        negated=True
                  else:
                    if list_operations_go[1] ==True and negated==False:
                      new_sentence= negate(sentence)
                      if new_sentence != sentence:
                          sentence= new_sentence
                          negated=True


        if noun_pertubation != '' and ent==False:
            tple = find_noun(sentence, sentence_parse) #(True, sentence, noun, conjugation)
            if tple[0] and tple[2]!='people':
                if negated:
                    noun_pertubation =make_nounList_with_hyper
                new_sentence = noun_pertubation (tple[2], tple[3], tple[1], sentence_parse)
                if new_sentence!=sentence:
                    sentence=new_sentence
                    if noun_pertubation =='make_nounList_with_anto':
                        negated=True
                else:
                  if list_operations_go[2]==True and negate ==False:        
                    new_sentence= Quantity_change(sentence, sentence_parse)
                    if new_sentence != sentence:
                        sentence= new_sentence
                        negated=True

        if list_operations_go[3]==True:   
            tple = find_adj(sentence, sentence_parse) #(True, sentence, adj, conjugation)
            if tple[0]:
                adj_list = make_adjList_with_anto(tple[2], tple[3])
                if adj_list!=[]:
                    np= editor.template(tple[1], adj = adj_list, remove_duplicates=True)
                    if np['data'] !=[]:
                        num= randrange(len(np['data']))
                        sentence = np['data'][num]
                        sentence_parse = parse([sentence])[1]
                    else:
                        sentence = sentence.replace("{adj}", tple[2]) 

                        
        if list_operations_go[4]==True and ent==False:
            ssplit=sentence.split()
            if len(ssplit)>2:
                sentence = roberta_add_word_before_noun_and_verb(sentence, sentence_parse)
                
                
        if list_operations_go[7]==True:
            new_sentence = hypo_or_other_premise(sentence, iter)
            if new_sentence != sentence:
                sentence= new_sentence

        if list_operations_go[8]==True:
            new_sentence = verb_past_to_future(sentence, sentence_parse)
            sentence_parse = new_sentence[1]
            if new_sentence[0] != sentence:
                sentence= new_sentence[0]

        if list_operations_go[9]==True:
            ssplit=sentence.split()
            if len(ssplit)>2:
                sentence = roberta_add_word_before_noun_and_verb(sentence, sentence_parse)

        if negated ==False:
            if list_operations_go[10]==True:
                new_sentence= negate(sentence)
                if new_sentence != sentence:
                    sentence= new_sentence
                    nagated=True
                else:
                  if list_operations_go[2]==True and negate ==False:        
                    new_sentence= Quantity_change(sentence, sentence_parse)
                    if new_sentence != sentence:
                        sentence= new_sentence
                        negated=True
               
        if sstart != sentence:
            old_premises.append(train_premises[iter][0])
            new_hypotheses.append(sentence)
            hyp.append(train_hypotheses[iter][0])
        else:
            if list_operations_go[6]==True:
                new_sentence=last_resort_negate(sentence)
                if new_sentence != sentence:
                    sentence= new_sentence

            if list_operations_go[0]==True:
                new_sentence = hypo_and_other_premise(sentence, iter)
                if new_sentence != sentence:
                    sentence= new_sentence
            
        print('-------------------------', iter , '--------------------------')
        iter=iter+1
    return (old_premises, new_hypotheses, hyp)

```

```python
def make_new_example(label, premise, premise_tok, hypothesis, hypothesis_tok):
    gold_label = label

    new_inference = {"annotator_labels": [label, label, label, label, label],
      "captionID": "6353564939.jpg#1"+" adapted to "+ label,
      "gold_label": gold_label,
      "pairID": "6353564939.jpg#1r1e" +" adapted to " + label,
      "sentence1": premise,
      "sentence1_binary_parse": "N/A",
      "sentence1_parse": premise_tok, 
      "sentence2": hypothesis ,
      "sentence2_binary_parse": "N/A",
      "sentence2_parse": hypothesis_tok}
    
    return new_inference
```

```python
def dump_in_file(filepath):
  with open(filepath, "a+") as outfile: 
      for new_example in ph_ent_to_neut:
          premise = new_example[1]
          premise_parse = nltk.word_tokenize(premise)
          hypothesis = new_example[2]
          hypothesis_parse = nltk.word_tokenize(hypothesis)
          label ="neutral"
          new_inference = make_new_example(label, premise, premise_parse, hypothesis, hypothesis_parse)
          json.dump(new_inference, outfile)
          outfile.write('\n')
      
      for new_example in ph_contr_to_neut:
          premise = new_example[1]
          premise_parse = nltk.word_tokenize(premise) 
          hypothesis = new_example[2]
          hypothesis_parse = nltk.word_tokenize(hypothesis) 
          label ="neutral"
          new_inference = make_new_example(label, premise, premise_parse, hypothesis, hypothesis_parse)
          json.dump(new_inference, outfile)
          outfile.write('\n')
          
      for new_example in ph_neut_to_ent:
          premise = new_example[1]
          premise_parse = nltk.word_tokenize(premise) 
          hypothesis = new_example[2]
          hypothesis_parse = nltk.word_tokenize(hypothesis) 
          label ="entailment"
          new_inference = make_new_example(label, premise, premise_parse, hypothesis, hypothesis_parse)
          json.dump(new_inference, outfile)
          outfile.write('\n')
          
      for new_example in ph_contr_to_ent:
          premise = new_example[1]
          premise_parse = nltk.word_tokenize(premise)
          hypothesis = new_example[2]
          hypothesis_parse = nltk.word_tokenize(hypothesis) 
          label ="entailment"
          new_inference = make_new_example(label, premise, premise_parse, hypothesis, hypothesis_parse)
          json.dump(new_inference, outfile)
          outfile.write('\n')
          
      for new_example in ph_neut_to_contr:
          premise = new_example[1]
          premise_parse = nltk.word_tokenize(premise)
          hypothesis = new_example[2]
          hypothesis_parse = nltk.word_tokenize(hypothesis)
          label ='contradiction'
          new_inference = make_new_example(label, premise, premise_parse, hypothesis, hypothesis_parse)
          json.dump(new_inference, outfile)
          outfile.write('\n')
          
      for new_example in ph_ent_to_contr:
          premise = new_example[1]
          premise_parse = nltk.word_tokenize(premise)
          hypothesis = new_example[2]
          hypothesis_parse = nltk.word_tokenize(hypothesis)
          label ='contradiction'
          new_inference = make_new_example(label, premise, premise_parse, hypothesis, hypothesis_parse)
          json.dump(new_inference, outfile)
          outfile.write('\n')
```
